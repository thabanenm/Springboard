These are your tasks for this mini project:

1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).
2. Convert the returned JSON object into a Python dictionary.
3. Calculate what the highest and lowest opening prices were for the stock in this period.
4. What was the largest change in any one day (based on High and Low price)?
5. What was the largest change between any two days (based on Closing Price)?
6. What was the average daily trading volume during this year?
7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
json_data = r.json() 
FSE_2017 = json_data.items()
for key, value in FSE_2017:
    print (key + ':',value)
```

```python
columns = json_data['dataset']['column_names']
data = json_data['dataset']['data']
df = pd.DataFrame(data, columns = columns)
df['Change'] = df['Open'] - df['Close']
df['Daily Turnover'] = df['Close'].shift(-1) - df['Close']
df.set_index('Date',inplace=True, drop=True)
FSE_dict = df.to_dict()
FSE_dict
```

```python
max_open = max(FSE_dict['Open'].values())
min_open = min(FSE_dict['Open'].values())
print('Highest opening price is ' + str(max_open) + ' and the lowest opening price is ' + str(min_open))
```

```python
max_change = max(FSE_dict['Change'].values())
print('The largest change in any one day is '+ str(max_change))
```

```python
max_diff = max(FSE_dict['Daily Turnover'].values())
print('The largest change between any two days is '+ str(max_diff))
```

```python
from statistics import mean
avg_trade = mean(FSE_dict['Traded Volume'].values())
print('The average daily trading volume during this year is ' + str(avg_trade))
```

```python
from statistics import median
median_trade = median(FSE_dict['Traded Volume'].values())
print('The median daily trading volume during this year is ' + str(median_trade))
```
